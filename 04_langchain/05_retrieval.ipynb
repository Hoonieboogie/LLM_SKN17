### Embedding Model

```python
from langchain_openai import OpenAIEmbeddings # hugging face 껄로 바꿀 수 있음

embedding_model = OpenAIEmbeddings(model='text-embedding-3-small') # 이거 HF Embedding이랑 repo로 바꾸면 됨

text = " ".join(documents[4].page_content.split()[1:10])
vector = embedding_model.embed_query(text)
print(vector)
print(len(vector))
```

```python
docs = [document.page_content for document in documents]
vects = embedding_model.embed_documents(docs)

print(len(vects))
print(len(vects[0]))
```

## Vector Store

```python
!pip install faiss-cpu
```

```python
from langchain.vectorstores import FAISS

text_embeddings = list(zip(docs, vects))
vector_store = FAISS.from_embeddings(text_embeddings, embedding_model)
#vector_store = FAISS.from_documents(documents, embedding_model) # 이렇게도 가능
```

```python
vector_store.save_local('./db/faiss') # 저장

# 불러오기
vector_db = FAISS.load_local(
    './db/faiss',
    embedding_model,
    allow_dangerous_deserialization=True
)

vector_db
```

```python
vector_store.similarity_search('Tom Sawyer', k=3)
```

### Retriever

```python
from langchain_openai import ChatOpenAI

retriever = vector_store.as_retriever()
print(retriever)

retriever.batch(['인디언 조'], k=3)
```

```python
from langchain.chains import RetrievalQA
from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    model = 'gpt-4o-mini',
    temperature=0
)

# 여기서 RAG를 구현한 것
retrieval_qa = RetrievalQA.from_chain_type(
    llm=model,
    retriever=retriever,
    chain_type='stuff'    
)
```

```python
retrieval_qa.invoke('인디언 조가 누구를 죽였나요')
```
