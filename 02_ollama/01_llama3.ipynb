# sLLM from Hugging Face

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = 'meta-llama/Llama-3.2-3B-Instruct' # granted by Hugging Face

tokenizer = AutoTokenizer.from_pretrained(model_name)
sllm = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    token=HF_TOKEN
)
```

```python
prompt = 'Intuitively explain what is quantum mechanics.'

inputs = tokenizer(prompt, return_tensors='pt') # Dict type return
outputs = sllm.generate(**inputs, max_length=500) # ** -> Thus, unpackaging
```

```python
print(inputs)
```

```python
print(outputs)
```

```python
# Decoding
print(tokenizer.decode(outputs[0], skip_special_tokens=True)) # 1) Show mt prompt 2) show response
```

### 3. 파라미터 조절

```python
import time
def generate_text_with_params(prompt, max_length=50, temperature=0.7, top_p=1):
    print(f"Prompt: {prompt} | max_length({max_length}), temperature({temperature}), top_p({top_p})")
    
    start_time = time.time()
    
    response = ollama.generate(
        model='llama3', 
        prompt=prompt,
        options={
            'max_length': max_length,
            'temperature': temperature,
            'top_p': top_p
        }
    )
    
    end_time = time.time()
    
    print(response['response'])
    print(f"Time passed: {end_time - start_time}")
    
```

```python
prompt = "오늘 점심 메뉴 한국어로 추천해줘"

max_lengths = [20, 50, 100]
temperatures = [0.3, 1.2]

for max_length in max_lengths:
    for temperature in temperatures:
        generate_text_with_params(prompt, max_length, temperature)
        print("=" * 100)
```

### 4. 다국어 처리 실험

```python
def generate_text_in_language(language, prompt):
    print(f'[언어: {language} | Prompt: {prompt}]')
    
    response = ollama.generate(model='llama3', prompt=prompt)
    print(response.response)
    print("="*100)
```

```python
prompts = {
    'English': "Recommend me one menu for today's lunch",
    '한국어': '오늘 점심 메뉴 하나만 한국어로 추천해줘',
    '日本語': '今日はランチメニューを1つだけ日本語でお勧めします。',
    '中文': '请用中文推荐一份今天的午餐菜单。',
    'Français': "S'il vous plaît, recommandez-moi un menu déjeuner aujourd'hui en français."
}

for lang, prompt in prompts.items():
    generate_text_in_language(lang, prompt)
```

### LLaMa 모델을 이용한 간단한 대화 시스템

- 사용자 입력 input() 이용
- 응답 print() 출력
- 'exit' 입력 시 종료

```python
import ollama

def ollama_llama_game():
    prompt = '[Always refer to the whole context of conversation when you respond].'
    while True:
        my_prompt = input('Write your prompt:')
        prompt = prompt + "Me: " + my_prompt + " "
        
        if my_prompt == 'exit':
            break
        
        print('You:', my_prompt)
        
        
        response = ollama.generate(model='llama3', prompt=prompt)


        print("Bot:", response.response)
        
        
        prompt = prompt + "Model: " + response.response + " "
        #print("------------------------------------------------------")
        #print(prompt)
        
        print()
```

```python
ollama_llama_game()
```
